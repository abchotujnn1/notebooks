* <u>**`target fields description`**</u>
* **cross_references:**

    - Description: Contains external references or links to the target in other databases or resources. These may include links to databases like
      UniProt, Ensembl, or other relevant biological databases.
    - Example: An array of dictionaries where each dictionary contains details about the external reference, such as the source and the external ID.

* organism:
    - Description: The scientific name of the organism from which the target is derived.
    - Example: "Homo sapiens" for human targets, "Mus musculus" for mouse targets.

* pref_name:
    - Description: The preferred name of the target. This is usually the most commonly used name for the target in scientific literature.
    - Example: "Beta-2 adrenergic receptor".

* score:
    - Description: This field typically represents a confidence score or relevance score associated with the target. The scoring system can vary and
     might indicate the reliability or significance of the target based on available data.
    - Example: 0.85 (Note: This field might not be present in all API responses).

* species_group_flag:
    - Description: A boolean flag indicating whether the target belongs to a species group, meaning it is a representative target for a group of     similar species.
    - Example: True or False.

* target_chembl_id:
    - Description: The unique identifier assigned to the target by ChEMBL. This ID is used to uniquely reference the target within the ChEMBL database.
    - Example: "CHEMBL210"

* target_components:
    - Description: Contains details about the components that make up the target. This is particularly relevant for complex targets like protein complexes or multi-subunit proteins.
    - Example: An array of dictionaries, where each dictionary includes information such as the component's description, accession numbers, and other relevant details.

* target_type:
    - Description: Describes the type or class of the target. Common target types include single proteins, protein complexes, protein families, and others.
    - Example: "SINGLE PROTEIN", "PROTEIN FAMILY".

* tax_id:
    - Description: The taxonomy ID of the organism from which the target is derived. This ID corresponds to the NCBI Taxonomy Database.
    - Example: 9606 for humans, 10090 for mice.
  
```json
{
  "cross_references": [
    {
      "xref_id": "P07550",
      "xref_name": "UniProtKB",

    }
  ],
  "organism": "Homo sapiens",
  "pref_name": "Beta-2 adrenergic receptor",
  "score": 0.85,
  "species_group_flag": false,
  "target_chembl_id": "CHEMBL210",
  "target_components": [
    {
      "accession": "P07550",
      "component_description": "Beta-2 adrenergic receptor",
      "component_id": 10548,
      "component_type": "PROTEIN"
    }
  ],
  "target_type": "SINGLE PROTEIN",
  "tax_id": 9606
}
```
* <u>**`Detailed Field Usage`**</u>
    - cross_references: Useful for integrating ChEMBL data with other biological databases.
    - organism: Essential for understanding the biological context of the target.
    - pref_name: Helpful for identifying the target by its common name.
    - score: Useful for prioritizing targets based on confidence or relevance.
    - species_group_flag: Important for understanding whether the target represents a broader group.
    - target_chembl_id: Key identifier for referencing and querying ChEMBL data.
    - target_components: Crucial for detailed structural and functional insights, especially for complex targets.
    - target_type: Provides context about the nature of the target, aiding in categorization and analysis.
    - tax_id: Facilitates integration with taxonomic and other biological datasets.

````python
bioactivity_data_coronavirus = bioactivity_data_coronavirus[bioactivity_data_coronavirus.standard_value.notna()]
````

````python
len(bioactivity_data_coronavirus)
````

````python
bioactivity_class = []
for i in bioactivity_data_coronavirus.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
````

````python
mol_cid = []
for i in bioactivity_data_coronavirus.molecule_chembl_id:
  mol_cid.append(i)
````

````python
canonical_smiles = []
for i in bioactivity_data_coronavirus.canonical_smiles:
  canonical_smiles.append(i)
````

````python
standard_value = []
for i in bioactivity_data_coronavirus.standard_value:
  standard_value.append(i)
````

````python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
df3.head()     
````

````python
import scanpy as sc
import pandas as pd
import GEOparse

# Define the GEO accession number for the dataset you want to fetch
geo_accession = "GSE130001"  # Replace with your specific accession number

# Fetch the GEO dataset
gse = GEOparse.get_GEO(geo=geo_accession, destdir="./")
print(gse.gsms.items())
# Combine all samples into a single DataFrame
data_frames = []
for gsm_name, gsm in gse.gsms.items():
    print("-------",gsm_name, gsm)
    print(gsm.table)
    data_frames.append(gsm.table)
print(data_frames)
# Assuming the first column is gene identifiers and the rest are counts
expression_data = pd.concat(data_frames, axis=1)
expression_data = expression_data.set_index(expression_data.columns[0])

# Create an AnnData object
adata = sc.AnnData(expression_data.T)

# Perform standard preprocessing steps
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata)
adata = adata[:, adata.var.highly_variable]

# Scale the data and run PCA
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata, svd_solver='arpack')

# Run UMAP and clustering
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)
sc.tl.leiden(adata)

# Visualize the data
sc.pl.umap(adata, color=['leiden'])
````

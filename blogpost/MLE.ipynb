```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate synthetic data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=30, scale=5, size=1000)
```

```python
def likelihood(parameters):
    mu, sigma = parameters
    pdf_values = norm.pdf(data, loc=mu, scale=sigma)
    return np.prod(pdf_values)

def log_likelihood(parameters):
    mu, sigma = parameters
    log_pdf_values = norm.logpdf(data, loc=mu, scale=sigma)
    return np.sum(log_pdf_values)
```

```python
from scipy.optimize import minimize

initial_guess = [0, 1]  # Initial guess for mean and standard deviation
result = minimize(lambda params: -log_likelihood(params), initial_guess)
estimated_params = result.x
```

```python
# Plot the data and the estimated distribution
plt.hist(data, bins=20, density=True, alpha=0.7, label='Data')
x_range = np.linspace(min(data), max(data), 2000)
estimated_pdf = norm.pdf(x_range, loc=estimated_params[0], scale=estimated_params[1])
plt.plot(x_range, estimated_pdf, color='red', label='Estimated PDF')
plt.title('Data and Estimated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

print("Estimated Mean:", estimated_params[0])
print("Estimated Standard Deviation:", estimated_params[1])
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate synthetic data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=30, scale=5, size=100)

# Define Likelihood and Log-Likelihood Functions
def likelihood(parameters):
    mu, sigma = parameters
    pdf_values = norm.pdf(data, loc=mu, scale=sigma)
    return np.prod(pdf_values)

def log_likelihood(parameters):
    mu, sigma = parameters
    log_pdf_values = norm.logpdf(data, loc=mu, scale=sigma)
    return np.sum(log_pdf_values)

# Maximize Log-Likelihood to Estimate Parameters
from scipy.optimize import minimize

initial_guess = [0, 1]  # Initial guess for mean and standard deviation
result = minimize(lambda params: -log_likelihood(params), initial_guess)
estimated_params = result.x

# Visualize the Results
plt.hist(data, bins=20, density=True, alpha=0.7, label='Data')
x_range = np.linspace(min(data), max(data), 100)
estimated_pdf = norm.pdf(x_range, loc=estimated_params[0], scale=estimated_params[1])
plt.plot(x_range, estimated_pdf, color='red', label='Estimated PDF')
plt.title('Data and Estimated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

print("Estimated Mean:", estimated_params[0])
print("Estimated Standard Deviation:", estimated_params[1])
```

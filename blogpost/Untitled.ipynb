```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, probplot

# Generate synthetic data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=30, scale=5, size=100)

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.tight_layout(pad=4.0)

# Histogram
sns.histplot(data, bins=15, kde=False, ax=axes[0, 0])
axes[0, 0].set_title('Histogram')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')

# Density Plot
sns.kdeplot(data, ax=axes[0, 1])
axes[0, 1].set_title('Density Plot')
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Density')

# Q-Q Plot
probplot(data, plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot')
axes[1, 0].set_xlabel('Theoretical Quantiles')
axes[1, 0].set_ylabel('Sample Quantiles')

# Q-Q Plot with Line
probplot(data, plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot with 45-degree Line')
axes[1, 1].set_xlabel('Theoretical Quantiles')
axes[1, 1].set_ylabel('Sample Quantiles')

plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
nsample = 100
rng = np.random.default_rng()

# A t distribution with small degrees of freedom:

ax1 = plt.subplot(221)
x = stats.t.rvs(3, size=nsample, random_state=rng)
res = stats.probplot(x, plot=plt)

# A t distribution with larger degrees of freedom:

ax2 = plt.subplot(222)
x = stats.t.rvs(25, size=nsample, random_state=rng)
res = stats.probplot(x, plot=plt)

# A mixture of two normal distributions with broadcasting:

ax3 = plt.subplot(223)
x = stats.norm.rvs(loc=[0,5], scale=[1,1.5],size=(nsample//2,2), random_state=rng).ravel()
res = stats.probplot(x, plot=plt)

# A standard normal distribution:

ax4 = plt.subplot(224)
x = stats.norm.rvs(loc=0, scale=1, size=nsample, random_state=rng)
res = stats.probplot(x, plot=plt)

# Produce a new figure with a loggamma distribution, using the ``dist`` and
# ``sparams`` keywords:

fig = plt.figure()
ax = fig.add_subplot(111)
x = stats.loggamma.rvs(c=2.5, size=500, random_state=rng)
res = stats.probplot(x, dist=stats.loggamma, sparams=(2.5,), plot=ax)
ax.set_title("Probplot for loggamma dist with shape parameter 2.5")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, probplot

# Generate synthetic data from two normal distributions
np.random.seed(42)
data_sample1 = np.random.normal(loc=30, scale=5, size=100)
data_sample2 = np.random.normal(loc=35, scale=7, size=100)

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.tight_layout(pad=4.0)

# Histograms for both samples
sns.histplot(data_sample1, bins=15, kde=False, color='blue', label='Sample 1', ax=axes[0, 0])
sns.histplot(data_sample2, bins=15, kde=False, color='orange', label='Sample 2', ax=axes[0, 0])
axes[0, 0].set_title('Histograms')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].legend()

# Density Plots for both samples
sns.kdeplot(data_sample1, color='blue', label='Sample 1', ax=axes[0, 1])
sns.kdeplot(data_sample2, color='orange', label='Sample 2', ax=axes[0, 1])
axes[0, 1].set_title('Density Plots')
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Density')
axes[0, 1].legend()

# Q-Q Plot for both samples
probplot(data_sample1, plot=axes[1, 0])
probplot(data_sample2, plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot')
axes[1, 0].set_xlabel('Theoretical Quantiles')
axes[1, 0].set_ylabel('Sample Quantiles')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data with different characteristics
np.random.seed(42)

# Continuous data with normal distribution
data_continuous = np.random.normal(loc=30, scale=5, size=1000)

# Discrete data with Poisson distribution
data_discrete = np.random.poisson(lam=10, size=100)

# Bimodal data with mixture of two normal distributions
data_bimodal = np.concatenate([
    np.random.normal(loc=25, scale=3, size=300),
    np.random.normal(loc=35, scale=3, size=700)
])

# Data with outliers
data_outliers = np.concatenate([
    np.random.normal(loc=30, scale=5, size=900),
    np.array([80, 90, 100])
])

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.tight_layout(pad=4.0)

# Plot data with different characteristics
axes[0, 0].hist(data_continuous, bins=20, color='blue', alpha=0.7)
axes[0, 0].set_title('Continuous Data')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')

axes[0, 1].hist(data_discrete, bins=20, color='orange', alpha=0.7)
axes[0, 1].set_title('Discrete Data')
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Frequency')

axes[1, 0].hist(data_bimodal, bins=20, color='green', alpha=0.7)
axes[1, 0].set_title('Bimodal Data')
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Frequency')

axes[1, 1].hist(data_outliers, bins=20, color='purple', alpha=0.7)
axes[1, 1].set_title('Data with Outliers')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Frequency')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Generate synthetic reaction time data (exponential distribution)
np.random.seed(42)
reaction_times = np.random.exponential(scale=0.5, size=1000)

# Fit the data to an exponential distribution
fit_params = expon.fit(reaction_times)

# Set up subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
fig.tight_layout(pad=4.0)

# Histogram of reaction times
axes[0].hist(reaction_times, bins=20, color='blue', alpha=0.7)
axes[0].set_title('Reaction Times (Exponential)')
axes[0].set_xlabel('Reaction Time')
axes[0].set_ylabel('Frequency')

# Fitted exponential PDF
x = np.linspace(0, max(reaction_times), 100)
pdf_fit = expon.pdf(x, *fit_params)
axes[1].plot(x, pdf_fit, color='green', label='Fitted Exponential')
axes[1].hist(reaction_times, density=True, bins=20, color='blue', alpha=0.5, label='Data Histogram')
axes[1].set_title('Fitted Exponential Distribution')
axes[1].set_xlabel('Reaction Time')
axes[1].set_ylabel('Probability Density')
axes[1].legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma, kstest, anderson, shapiro

# Generate synthetic data from a gamma distribution
data = np.random.gamma(shape=2, scale=1, size=1000)

# Perform Kolmogorov-Smirnov test
ks_stat, ks_p_value = kstest(data, 'norm')

# Perform Anderson-Darling test
ad_stat, ad_critical_values, ad_significance_levels = anderson(data, 'norm')

# Perform Shapiro-Wilk test
shapiro_stat, shapiro_p_value = shapiro(data)

# Plot the histogram of data and fitted distributions
plt.hist(data, density=True, bins=20, alpha=0.5, label='Data Histogram')

# Fit normal distribution
mu, std = norm.fit(data)
x = np.linspace(min(data), max(data), 100)
pdf_normal = norm.pdf(x, mu, std)
plt.plot(x, pdf_normal, label='Fitted Normal')

# Fit gamma distribution
a, loc, scale = gamma.fit(data)
pdf_gamma = gamma.pdf(x, a, loc, scale)
plt.plot(x, pdf_gamma, label='Fitted Gamma')

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

# Print test results
print("KS Test (Normal):", ks_p_value)
print("AD Test (Normal):", ad_stat)
print("Shapiro-Wilk Test:", shapiro_p_value)
```
